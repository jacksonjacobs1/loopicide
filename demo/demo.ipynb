```python
import numpy as np
import matplotlib.pyplot as plt
from loopicide import remove_loops
```

```python
problem_example = np.array([(0,0), (0,-1), (0,-3), (1,-1), (1,-3), (2,-2), (-1,-2),(-3,-2),(-2, -3), (-2, 0)])
problem_example_no_loops = remove_loops(problem_example)
plt.plot(problem_example[:,0], problem_example[:,1], '--b', marker='o', label='Original')
plt.plot(problem_example_no_loops[:,0], problem_example_no_loops[:,1], '-r', marker='.', label='No loops')
plt.legend()
plt.show()
```
